```python
def IQR(data):
    IQR=data.quantile(.75)-data.quantile(.25)
    return IQR
```

```python
a=IQR(df['AverageWinRate'])
```

```python
def upperBound(data):
    upperBound=data.quantile(.75)+1.5*IQR(data)
    return upperBound
    
```

```python
def lowerBound(data):
    lowerBound= data.quantile(.25)-1.5*IQR(data)
    return lowerBound
```

```python
outRe=df['AverageKill'].between(lowerBound(df['AverageKill']), upperBound(df['AverageKill']), inclusive=True)
```
